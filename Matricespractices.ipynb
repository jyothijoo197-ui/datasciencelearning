```python
import numpy as np

mat1 = np.array([
    [1,2],
    [3,4]
])
mat2 = np.array([
    [5,6],
    [7,8]
])
result = mat1 + mat2
print(result)
```

```python
#subtraction
mat1 = np.array([
    [5,6],
    [7,8]
])
mat2 =np.array([
    [1,2],
    [3,4]
])
result = mat1 - mat2
print(result)
```

```python
#multiplication
mat1 = np.array([
    [1,2],
    [3,4]
])
mat2 = np.array([
    [5,6],
    [7,8]
])
result = mat1*mat2
print(result)
```

```python
#matrix multiplication
mat1 = np.array([
    [1,2],
    [3,4]
])
mat2 = np.array([
    [5,6],
    [7,8]
])
result = np.matmul(mat1,mat2)
print(result)
result = mat1 @mat2
print(result)
result = np.dot(mat1,mat2)
print(result)
```

```python
#division
mat1 = np.array([
    [10,20],
    [30,40]
])
mat2 = np.array([
    [5,6],
    [7,8]
])
result = mat1/mat2
print(result)
```
