```python
import pandas as pd
import numpy as np
```

```python
#series
arr = np.array([10,20,30,40,50])
marks = pd.Series(arr)
print(marks)
print(marks[0])
```

```python
marks = pd.Series(
    [10,20,30],
    index = ["math","phy","chem"]
)
print(marks)
print(marks["math"])
print(marks.shape)
print(marks.size)
print(marks.dtype)
print(marks.values)
print(marks.index)
```

```python
# => 2,4,68,14
numbers = [1,2,34,7]
result = []
for i in numbers:
  result.append(i*2)
print(result)
```

```python
# vectorized opeartion means applying one opeartion to all elements automatically without writing loops
numbers = [1,2,34,7]
arr = np.array(numbers)
print(arr*2)
```

```python
data = pd.Series([10,20,30,40, np.nan, 50])
# [10,20,30,40, np.nan, 50]
print(data.isnull()) # if i want to know the missing values i will use  is null method
print(data.fillna(10))
```

```python
#dataframe
data = {
    "name": ["lahari", "jagadeesh", "jyothikiran", "swathi"],
    "marks": [60,61,70,71],
    "city": ["vij", "bang", "bangkok","london"]
}
df = pd.DataFrame(data)
print(df)
print(df.shape)
print(df.size)
print(df.columns)
print(df.dtypes)
print(df.index)
```

```python
df["Grade"] = ["B", "B", "A","A"]
print(df)
```

```python
df["marks"] = df["marks"] + 20
print(df)
df.drop("city", axis=1)
```
